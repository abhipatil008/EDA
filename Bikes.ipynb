```python
msell=df['selling_price'].mean()
mesell=df['selling_price'].median()
sns.distplot(df['selling_price'])
plt.axvline(msell,color='r')
plt.axvline(mesell,color='y')
plt.grid(True)
```

```python
#As the mean of the selling_price is greater than median it is positively skewed
```
